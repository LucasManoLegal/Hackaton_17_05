```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_produtos = pd.read_excel("C:/Users/49439967870/Desktop/Ciência_de_dados/Hackaton_17_05/tratamento/vendas.xlsx")
print(df_produtos)
```

```python
df_produtos["Preco_unitario"] = df_produtos["Preco_unitario"].astype(float)
display(df_produtos)
```

```python
df_produtos["Data_venda"] = pd.to_datetime(df_produtos["Data_venda"],unit = "ms")
df_produtos["Data_venda"] = df_produtos["Data_venda"].dt.strftime('%d/%m/%y')
display(df_produtos)
```

```python
df_produtos['Pedidos_concluido'].fillna(0, inplace=True)

def ehnulo(linha):
    import math
    if math.isnan(linha):
        return "Nao Concluido"
    else:
        return "Concluido"

df_produtos["Pedidos_concluido"] = df_produtos["Avaliacao_servico"].apply(ehnulo)
       
        

print(df_produtos)
```

```python
df_produtos["Valor_total"] = df_produtos["Valor_total"].astype(float)
display(df_produtos)
```
